```python
import pandas as pd
df = pd.read_csv('C:/Users/Jake/Jake_Dennis_DS-Unit-2-Applied-Modeling/data/fantasy3.csv')
print(df.shape)
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['PosRank', 'Rec', 'FantPos'])
print(df.shape)
df.isnull().sum()
```

```python
train = df[df['Year'] < 2016]
test = df[df['Year'] > 2015]
val = train[train['Year'] > 2013]
```

```python
import numpy as np
target = ['PosRank']
drop = ['Year', 'Name', 'PosRank']
X_train = train.drop(columns=drop)
y_train = train[target]
X_val = val.drop(columns=drop)
y_val = val[target]
X_test = test.drop(columns=drop)
```

```python
print(y_train.shape)
X_train.shape
```

```python
X_test.head()
```

```python
import category_encoders as ce
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from scipy.stats import randint, uniform
from sklearn.model_selection import RandomizedSearchCV

pipeline = make_pipeline(
    ce.TargetEncoder(), 
    SimpleImputer(), 
    RandomForestRegressor(random_state=42)
)

pipeline.fit(X_train, y_train.values.ravel())
print ('Validation Accuracy', pipeline.score(X_val, y_val))
```

```python
test_pred = pipeline.predict(X_test)
```

```python
print(pipeline.score(X_test, test_pred))
```
